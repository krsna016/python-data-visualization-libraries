```python
# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
# Note: Also add title, xlabel, and ylabel to the plot.

plt.scatter(x, y, color='blue')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

# The `subplot()` function is used to create multiple plots in a single figure. It divides the plotting area into a grid and allows plotting multiple graphs in separate sub-regions of the grid.
# Four line plots:
import matplotlib.pyplot as plt
import numpy as np

# Data for the plots
x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Creating subplots
plt.subplot(2, 2, 1)
plt.plot(x, y1, label='Line 1', color='red')
plt.title('Line 1')

plt.subplot(2, 2, 2)
plt.plot(x, y2, label='Line 2', color='blue')
plt.title('Line 2')

plt.subplot(2, 2, 3)
plt.plot(x, y3, label='Line 3', color='green')
plt.title('Line 3')

plt.subplot(2, 2, 4)
plt.plot(x, y4, label='Line 4', color='purple')
plt.title('Line 4')

plt.tight_layout()
plt.show()
```

```python
#  Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

# A bar plot is a graph that represents categorical data using rectangular bars. The length of each bar corresponds to the value of the category it represents. Bar plots are used to compare values across categories.
# Bar Plot and Horizontal Bar Plot:
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Vertical Bar Plot
plt.bar(company, profit, color='orange')
plt.title('Company Profit')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.show()

# Horizontal Bar Plot
plt.barh(company, profit, color='green')
plt.title('Company Profit')
plt.xlabel('Profit')
plt.ylabel('Company')
plt.show()
```

```python
# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

# A box plot is a statistical plot that represents the distribution of a dataset through its quartiles. It shows the minimum, first quartile (Q1), median, third quartile (Q3), and maximum values. Box plots are used to identify outliers and understand the spread of data.


import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.boxplot([box1, box2], tick_labels=['Box1', 'Box2'], patch_artist=True)
plt.title('Box Plot')
plt.ylabel('Values')
plt.show()
```
