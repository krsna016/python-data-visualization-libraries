```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,200)
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel("This is my X data")
plt.ylabel("This is my Y data")
plt.title("This is my Title")
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
x,y
```

```python
plt.plot(x,y)
```

```python
plt.scatter(x,y)
```

```python
colours = np.random.rand(50)
sizes = np.random.rand(50)
plt.scatter(x,y,c=colours,s=sizes*1000, alpha=0.5)
```

```python
x = ['a','b','c','d','e'] # Categorical value
y = np.random.rand(5) # Numerical value
plt.bar(x,y)
```

```python
plt.barh(x,y)
plt.xlabel('This is X-Label')
plt.ylabel('This is Y-Label')
plt.title('This is Title')
plt.show()
```

```python
x = np.linspace(0,10,200)
y = np.sin(x)
plt.plot(x,y,'--g')
```

```python
x = np.linspace(0,10,200)
y = np.sin(x)
plt.figure(figsize=(10,2))
plt.grid()
plt.plot(x,y)
```
